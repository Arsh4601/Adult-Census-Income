```python
from sklearn.preprocessing import LabelEncoder
```

```python
lenc=LabelEncoder()
```

```python
data1
```

```python
data1.columns
```

```python
data1["income"].value_counts()
```

```python
data1["income"].replace(["<=50K"],0,inplace=True)
data1["income"].replace([">50K"],1,inplace=True)
data1
```

```python
data1["workclass"]=lenc.fit_transform(data1["workclass"])
data1['education']=lenc.fit_transform(data1['education'])
data1['marital.status']=lenc.fit_transform(data1['marital.status'])
data1['relationship']=lenc.fit_transform(data1['relationship'])
data1['occupation']=lenc.fit_transform(data1['occupation'])
data1["race"]=lenc.fit_transform(data1["race"])
data1["sex"]=lenc.fit_transform(data1["sex"])
data1["native.country"]=lenc.fit_transform(data1["native.country"])

data1
```

```python
plt.figure(figsize=(16,9))
corr=data1.corr() 

sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns,annot=True)
plt.show()
```
