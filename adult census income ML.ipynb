# Data Analysis

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix,classification_report
```

```python
data=pd.read_csv("C:/Users/Shafiq/Desktop/arsh LU/adult income/adult_cleanedenc.csv")

data
```

# Treating imbalance

```python
x=data.drop("income",axis=1)
y=data["income"]
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE(sampling_strategy="minority",random_state=0)
```

```python
x_sm,y_sm=sm.fit_resample(x,y)
```

```python
y_sm.value_counts()
```

```python
x_sm
```

```python
data_balance=x_sm.copy()
data_balance["income"]=y_sm
data_balance
```

# Feature Selection on basis of data analysis

```python
data_balance.columns
```

```python
x=data_balance.drop(["income","workclass","fnlwgt","native.country"],axis=1)
y=data_balance["income"]
```

```python
x
```

```python
y
```

```python
y.value_counts()
```

# Preparing and Scaling train and test dataset

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
print("x_train shape: ",x_train.shape)
print("y_train shape: ",y_train.shape)
print("x_test shape: ",x_test.shape)
print("y_test shape: ",y_test.shape)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

# Scaling x_train and x_test

```python
from sklearn.preprocessing import StandardScaler
sc1=StandardScaler()
sc2=StandardScaler()
```

```python
x_test=sc1.fit_transform(x_test)
x_train=sc2.fit_transform(x_train)
```

```python
x_train
```

```python
x_test
```

# ML Modelling

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(x_train,y_train)
y_train_pred=logreg.predict(x_train)
y_test_pred=logreg.predict(x_test)

print("Accuracy train: ",accuracy_score(y_train,y_train_pred))
print("confusion_matrix train:\n",confusion_matrix(y_train,y_train_pred))
print("classification_report train:\n",classification_report(y_train,y_train_pred))
print("Precision train: ",precision_score(y_train,y_train_pred),"\n")

print("Accuracy test: ",accuracy_score(y_test,y_test_pred))
print("confusion_matrix test:\n",confusion_matrix(y_test,y_test_pred))
print("classification_report test:\n",classification_report(y_test,y_test_pred))
print("Precision test: ",precision_score(y_test,y_test_pred))
```

# Logreg not working very vell trying Hyperparamater Tuning

# Hyperparamater Tuning

```python
for c in np.logspace(-5,8,15):
    
    print("at C: ",c,"\n")
    logreg=LogisticRegression(C=c)
    logreg.fit(x_train,y_train)
    y_train_pred=logreg.predict(x_train)
    y_test_pred=logreg.predict(x_test)
    
    print("Accuracy train: ",accuracy_score(y_train,y_train_pred))
    print("Precision train: ",precision_score(y_train,y_train_pred),"\n")
    print("Accuracy test: ",accuracy_score(y_test,y_test_pred))
    print("Precision test: ",precision_score(y_test,y_test_pred),"\n")   
    
```
